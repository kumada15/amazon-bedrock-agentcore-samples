## ã‚¹ãƒ†ãƒƒãƒ— 4: Invoke é–¢æ•°ã®ä½œæˆ

````python
def invoke(payload):
    """ã‚¨ãƒ¼ã‚¸ã‚§ãƒ³ãƒˆæ“ä½œã®ãŸã‚ã®æ§‹é€ åŒ–ã•ã‚ŒãŸ invoke é–¢æ•°"""
    user_message = payload.get("prompt", "")
    
    try:
        print("åˆ†æã‚’é–‹å§‹ã—ã¦ã„ã¾ã™...")
        start_time = time.time()
        
        response = strands_agent(user_message)
        
        elapsed_time = time.time() - start_time
        print(f"åˆ†æãŒ {elapsed_time:.2f} ç§’ã§å®Œäº†ã—ã¾ã—ãŸ")
        
        return response.message["content"][0]["text"] if response.message.get("content") else str(response)
        
    except Exception as e:
        print(f"åˆ†æä¸­ã«ã‚¨ãƒ©ãƒ¼ãŒç™ºç”Ÿã—ã¾ã—ãŸ: {str(e)}")
        return f"ã‚¨ãƒ©ãƒ¼ãŒç™ºç”Ÿã—ã¾ã—ãŸ: {str(e)}"

print("Invoke é–¢æ•°ãŒä½œæˆã•ã‚Œã¾ã—ãŸï¼")

# æ³¨æ„ï¼šã‚¿ã‚¤ãƒ ã‚¢ã‚¦ãƒˆã‚’å¼•ãèµ·ã“ã™é…ã„ Web ã‚µã‚¤ãƒˆã«ã‚¢ã‚¯ã‚»ã‚¹ã™ã‚‹å ´åˆã¯ã€
# ãƒãƒ³ã‚°ã™ã‚‹æ“ä½œã‚’é˜²ããŸã‚ã«ã‚³ãƒ¼ãƒ‰ã«ã‚¿ã‚¤ãƒ ã‚¢ã‚¦ãƒˆå‡¦ç†ã‚’å®Ÿè£…ã™ã‚‹å¿…è¦ãŒã‚ã‚‹å ´åˆãŒã‚ã‚Šã¾ã™ã€‚
````

````python
def analyze_website(url, question=None):
    """ã‚¤ãƒ³ã‚¿ãƒ©ã‚¯ãƒ†ã‚£ãƒ–ãª Web ã‚µã‚¤ãƒˆåˆ†æã®ãŸã‚ã®ä¾¿åˆ©ãªé–¢æ•°"""
    if question:
        prompt = f"{url} ã«ã‚¢ã‚¯ã‚»ã‚¹ã—ã¦æ¬¡ã®è³ªå•ã«ç­”ãˆã¦ãã ã•ã„: {question}"
    else:
        prompt = f"{url} ã® Web ã‚µã‚¤ãƒˆã‚’åˆ†æã—ã€åŒ…æ‹¬çš„ãªè²¡å‹™ã‚¤ãƒ³ã‚µã‚¤ãƒˆã‚’æä¾›ã—ã¦ãã ã•ã„"
    
    payload = {"prompt": prompt}
    return invoke(payload)

# ä½¿ç”¨ä¾‹
print("ã‚¤ãƒ³ã‚¿ãƒ©ã‚¯ãƒ†ã‚£ãƒ–åˆ†æé–¢æ•°ã®æº–å‚™ãŒã§ãã¾ã—ãŸï¼")
print("\nä½¿ç”¨ä¾‹:")
print('analyze_website("https://www.marketwatch.com/investing/stock/aapl", "Apple ã®ç¾åœ¨ã®æ ªå¼ãƒ‘ãƒ•ã‚©ãƒ¼ãƒãƒ³ã‚¹ã¯ï¼Ÿ")')
print("\nã‚¤ãƒ³ã‚¿ãƒ©ã‚¯ãƒ†ã‚£ãƒ–åˆ†æé–¢æ•°ã®æº–å‚™å®Œäº†ï¼")
````
