### ステップ 2: フロントエンドアーキテクチャの理解

Streamlit アプリケーションは、いくつかの主要なコンポーネントで構成されています：

#### コアコンポーネント:

1. **main.py** - UI と認証を備えたメイン Streamlit アプリケーション
2. **chat.py** - チャット管理と AgentCore Runtime 統合
3. **chat_utils.py** - メッセージのフォーマットと表示のためのユーティリティ関数
4. **sagemaker_helper.py** - アクセス可能な URL 生成のためのヘルパー

#### 認証フロー:

1. ユーザーが Streamlit アプリケーションにアクセス
2. Amazon Cognito がユーザー認証を処理
3. 有効な JWT トークンが AgentCore Runtime リクエストの認可に使用される
4. ユーザーはカスタマーサポートエージェントと安全にやり取りできる
